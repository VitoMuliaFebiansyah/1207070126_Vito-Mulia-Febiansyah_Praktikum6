```python
import numpy as np #import modul numpy sebagai np
import cv2 #import modul cv2
import matplotlib.pyplot as plt #import modul matplotlib.pyplot sebagai plt
```

```python
img = cv2.imread("D:\Semester 6\PCD\cod.jpg") #untuk membaca gambar yang ada pada folder semester 6 -> PCD -> cod.jpg dan menyimpannya pada variabel img
```

```python
img_height = img.shape[0] #untuk mendapatakan informasi nilai tinggi dari gambar dan disimpan pada variabel img_height
img_width = img.shape[1] #untuk mendapatkan informasi nilai lebar dari gambar dan disimpan pada variabel img_width
img_channel = img.shape[2] #untuk mendapatkan informasi jumlah channel dari gambar dan disimpan pada variabel img_channel
img_type = img.dtype #untuk mendapatkan informasi tipe data dari gambar, seperti uint8 atau float32 dan disimpan pada variabel img_type
```

```python
img_flip_horizontal = np.zeros(img.shape, img_type) #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan memanggil variabel img_type dan disimpan pada variabel img_flip_horizontal
img_flip_vertical = np.zeros(img.shape, img_type) #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan memanggil variabel img_type dan disimpan pada variabel img_flip_vertical
```

```python
for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
    for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
        for c in range(0, img_channel): #untuk melakukan loop pada setiap nilai c
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] #setiap pixel pada (y,x) akan ditukar posisinya dengan [img_width-1-x] ke posisi channel c dan disimpan pada variabel img_flip_horizontal
```

```python
for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
    for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
        for c in range(0, img_channel): #untuk melakukan loop pada setiap nilai c
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c] #setiap pixel pada (y,x) akan ditukar posisinya dengan [img_height-1-y] ke posisi channel c dan disimpan pada variabel img_flip_vertical
```

```python
plt.imshow(img_flip_horizontal) #menampilkan gambar pada variabel img_flip_horizontal
plt.title("Flip Horizontal") #diberi judul Flip Horizontal
plt.show() #untuk menampilkan output pada layar
plt.imshow(img_flip_vertical) #menampilkan gambar pada variabel img_flip_vertical
plt.title("Flip Vertical") #diberi judul Flip Vertical
plt.show() #untuk menampilkan output pada layar
```
