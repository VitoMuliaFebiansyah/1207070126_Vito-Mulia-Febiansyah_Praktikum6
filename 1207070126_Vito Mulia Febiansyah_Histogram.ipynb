```python
import numpy as np #import modul numpy sebagai np
import cv2 #import modul cv2
import matplotlib.pyplot as plt #import modul matplotlib.pyplot sebagai plt
```

```python
img = cv2.imread("D:\Semester 6\PCD\cod.jpg") #untuk membaca gambar yang ada pada folder semester 6 -> PCD -> cod.jpg dan menyimpannya pada variabel img
```

```python
img_height = img.shape[0] #untuk mendapatakan informasi nilai tinggi dari gambar dan disimpan pada variabel img_height
img_width = img.shape[1] #untuk mendapatkan informasi nilai lebar dari gambar dan disimpan pada variabel img_width
img_channel = img.shape[2] #untuk mendapatkan informasi jumlah channel dari gambar dan disimpan pada variabel img_channel
```

```python
img_grayscale = np.zeros(img.shape, dtype=np.uint8)  #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan tipe data uint8 dan disimpan pada variabel img_grayscale

for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
    for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
        red = img[y][x][0] #mengambil nilai red pada pixel (y,x) dari gambar dan disimpan pada variabel red
        green = img[y][x][1] #mengambil nilai green pada pixel (y,x) dari gambar dan disimpan pada variabel green
        blue = img[y][x][2] #mengambil nilai blue pada pixel (y,x) dari gambar dan disimpan pada variabel blue
        gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
        img_grayscale[y][x] = (gray, gray, gray) #nilai keabuan yang telah diubah pada setiap pixel (y,x) disimpan pada variabel img_grayscale
        
plt.imshow(img_grayscale) #menampilkan gambar pada variabel img_grayscale
plt.title("Grayscale") #diberi judul Grayscale
plt.show() #untuk mena,pilkan output di layar
```

```python
hg = np.zeros((256)) #membuat sebuah array satu dimensi dengan panjang 256 yang berisi nol dan disimpan pada variabel hg
```

```python
for x in range(0, 256): #untuk melakukan loop pada setiap nilai x dari 0-255
    hg[x] = 0 #inisialisasi nilai 0 pada variabel hg untuk setiap elemen array hg
```

```python
for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
    for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
        gray = img_grayscale[y][x][0] #mengambil nilai pixel grayscale (y,x)
        hg[gray] += 1 #menambahkan 1 pada setiap elemen hg
```

```python
bins = np.linspace(0, 256, 100) #untuk membuat array berisi nilai-nilai yang secara merata tersebar dalam rentang dari 0 hingga 255, dengan jumlah elemen sebanyak 100
plt.hist(hg, bins, color="black", alpha=0.5) #membuat histogram dari gambar grayscale
plt.title("Histogram") #diberi judul Histogram
plt.show() #untuk menampilkan output di layar
```

```python
hgk = np.zeros((256)) #membuat sebuah array satu dimensi dengan panjang 256 yang berisi nol dan disimpan pada variabel hgk
c = np.zeros((256)) #membuat sebuah array satu dimensi dengan panjang 256 yang berisi nol dan disimpan pada variabel c

for x in range(0, 256): #untuk melakukan loop pada setiap nilai x dari 0-255
    hgk[x] = 0 #inisialisasi nilai 0 pada hgk
    c[x] = 0 #inisialisai nilai 0 pada c

for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
    for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
        gray = img_grayscale[y][x][0] #mengambil nilai pixel grayscale (y,x)
        hgk[gray] += 1 #menambahkan nilai 1 pada hgk
                
c[0] = hgk[0] #inisialisasi dari array kumulatif dengan nilai histogram pixel pertama
for x in range(1, 256): #untuk melakukan loop pada setiap nilai x dari 1-255
     c[x] = c[x-1] + hgk[x] #operasi pertambahan antara variabel c dan hgk

hmaxk = c[255] #menyimpan nilai kumulatif maksimum pada variabel hmaxk

for x in range(0, 256): #untuk melakukan loop pada setiap nilai x dari 0-255
    c[x] = 190 * c[x] / hmaxk #mengalikan setiap nilai dengan 190 dan membaginya dengan hmaxk

plt.hist(c, bins, color="black", alpha=0.5) #untuk membuat histogram dari c
plt.title("Histogram Grayscale Kumulatif") #diberi judul Histogram Grayscale Kumulatif
plt.show() #untuk menampilkan output di layar
```

```python
hgh = np.zeros((256)) #membuat sebuah array satu dimensi dengan panjang 256 yang berisi nol dan disimpan pada variabel hgh
h = np.zeros((256)) #membuat sebuah array satu dimensi dengan panjang 256 yang berisi nol dan disimpan pada variabel h
c = np.zeros((256)) #membuat sebuah array satu dimensi dengan panjang 256 yang berisi nol dan disimpan pada variabel c

for x in range(0, 256): #untuk melakukan loop pada setiap nilai x dari 0-255
    hgh[x] = 0 #inisialisasi nilai 0 pada hgh
    h[x] = 0 #inisialisasi nilai 0 pada h
    c[x] = 0 #inisialisasi nilai 0 pada c

for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y 
    for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
        gray = img_grayscale[y][x][0] #mengambil nilai pixel grayscale (y,x)
        hgh[gray] += 1 #menambahkan nilai 1 pada hgh 
                
h[0] = hgh[0] #inisialisasi dari array hequalisasi dengan nilai histogram pixel pertama
for x in range(1, 256): #untuk melakukan loop pada setiap x dari 1-255
     h[x] = h[x-1] + hgh[x] #operasi pertambahan antara h dan hgh serta disimpan pada variabel h

for x in range(0, 256): #untuk melakukan loop pada setiap nilai x dari 0-255
     h[x] = h[x] / img_height / img_width #operasi pembagian untuk nilai h

for x in range(0, 256): #untuk melakukan loop pada setiap nilai x dari 0-255
    hgh[x] = 0 #inisialisasi nilai 0 pada hgh
    
for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
    for x in range(0, img_width): #untuk melakukan loop pada setuap nilai x
        gray = img_grayscale[y][x][0] #mengambil nilai pixel grayscale (y,x)
        gray = h[gray] * 255 #mengalikan nilai h dengan 255
        hgh[int(gray)] += 1 #menambahkan nilai 1 pada hgh menjadi interger

c[0] = hgh[0] #inisialisasi dari array hequalisasi dengan nilai histogram pixel pertama
for x in range(1, 256): #untuk melakukan loop pada setiap x dari 1-255
     c[x] = c[x-1] + hgh[x] #operasi pertambhan antara c dan hgh

hmaxk = c[255] #menyimpan nilai maksimum pada variabel hmaxk

for x in range(0, 256): #untuk melakukan loop pada setiap nilai c dari 0-255
    c[x] = 190 * c[x] / hmaxk #mengalikan setiap nilai dengan 190 dan membaginya dengan hmaxk

plt.hist(c, bins, color="black", alpha=0.5) #untuk membuat histogram hequalisasi
plt.title("Histogram Grayscale Hequalisasi") #diberi judul Histogram Grayscale Hequalisasi
plt.show() #untuk menampilkan output di layar
```
