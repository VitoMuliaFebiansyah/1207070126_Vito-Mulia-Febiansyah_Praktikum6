```python
import numpy as np #import modul numpy sebagai np
import cv2 #import modul cv2
import matplotlib.pyplot as plt #import modul matplotlib.pyplot sebagai plt
```

```python
img = cv2.imread("D:\Semester 6\PCD\cod.jpg") #untuk membaca gambar yang ada pada folder semester 6 -> PCD -> cod.jpg dan menyimpannya pada variabel img
```

```python
img_height = img.shape[0] #untuk mendapatakan informasi nilai tinggi dari gambar dan disimpan pada variabel img_height
img_width = img.shape[1] #untuk mendapatkan informasi nilai lebar dari gambar dan disimpan pada variabel img_width
img_channel = img.shape[2] #untuk mendapatkan informasi jumlah channel dari gambar dan disimpan pada variabel img_channel
```

```python
img_inversi = np.zeros(img.shape, dtype=np.uint8)  #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan tipe data uint8 dan disimpan pada variabel img_inversi
```

```python
def inversi_grayscale(nilai): #mendefinisikan inversi_grayscale dengan parameter nilai
    for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
        for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
            red = img[y][x][0] #mengambil nilai red dari pixel (y,x) pada gambar
            green = img[y][x][1] #mengambil nilai green dari pixel (y,x) pada gambar
            blue = img[y][x][2] #mengambil nilai blue dari pixel (y,x) pada gambar
            gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
            gray = nilai - gray #menghitung selisih dengan cara mengurangkan variabel gray dengan nilai
            img_inversi[y][x] = (gray, gray, gray) #nilai keabuan yang telah diubah pada setiap pixel (y,x) disimpan pada variabel img_inversi
```

```python
def inversi_rgb(nilai): #mendefinisikan inversi_rgb dengan parameter nilai
    for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
        for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
            red = img[y][x][0] #mengambil nilai red dari pixel (y,x) pada gambar
            red = nilai - red #menghitung selisih dan disimpan pada variabel red
            green = img[y][x][1] #mengambil nilai red dari pixel (y,x) pada gambar 
            green = nilai - green #menghitung selisih atau operasi pengurangan 
            blue = img[y][x][2] #mengambil nilai blue dari pixel (y,x) pada gambar
            blue = nilai - blue #operasi pengurangan antar nilai dan blue
            img_inversi[y][x] = (red, green, blue) #nilai RGB yang telah diubah pada setiap pixel (y,x) disimpan pada variabel img_inversi
```

```python
inversi_grayscale(255) #menerapkan inversi_grayscale dengan mengurangkan pixel dari nilai maksimum 255
plt.imshow(img_inversi) #menampilkan gambar dari img_inversi
plt.title("Inversi Grayscale") #diberi judul Inversi Grayscale
plt.show() #untuk menampilkan output di layar

inversi_rgb(255) #menerapkan inversi_rgb dengan mengurangkan pixel dari nilai maksimum 255
plt.imshow(img_inversi) #menampilkan gamabr dari img_inversi
plt.title("Inversi RGB") #diberi judul Inversi RGB
plt.show() #untuk menampilkan output di layar
```

```python
img_log = np.zeros(img.shape, dtype=np.uint8) #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan tipe data uint8 dan disimpan pada variabel img_log
```

```python
def log(c): #mendefinisikan log dengan parameter c
    for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
        for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
            red = img[y][x][0] #mengambil nilai red dari pixel (y,x) pada gambar
            green = img[y][x][1] #mengambil nilai green dari pixel (y,x) pada gambar
            blue = img[y][x][2] #mengambil nilai blue dari pixel (y,x) pada gambar
            gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
            gray = int(c * np.log(gray + 1)) #mengambil nilai gray dari pixel (y,x), lalu melakukan operasi logaritmik pada nilai tersebut
            if gray > 255: #mengecek nilai gray jika lebih dari 255
                gray = 255 #jika lebih dari 255, maka nilai gray akan diset jadi 255
            if gray < 0: #mengecek nilai gray jika kurang dari 0
                gray = 0 #jika kurang dari 0, maka nilai gray akan diset jadi 0
            img_log[y][x] = (gray, gray, gray) #nilai keabuan yang telah diubah pada setiap pixel (y,x) disimpan pada variabel img_log
```

```python
log(30) #menerapkan operasi logaritma dengan nilai log 30 
plt.imshow(img_log) #menampilkan gambar dari img_log
plt.title("Log") #diberi judul Log
plt.show() #untuk menampilkan output di layar
```

```python
img_nthpower = np.zeros(img.shape, dtype=np.uint8) #untuk membuat array numpy kosong dengan dimensi yang sama seperti array img dengan tipe data uint8 dan disimpan pada variabel img_nthpower
```

```python
def nthpower(c, y): #mendefinisikan nthpower dengan parameter c dan y
    thc = c / 100 #inisialisasi nilai c dibagi 100 untuk variabel thc
    thy = y / 100 #inisialisasi nilai y dibagi 100 untuk variabel thy
    for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
        for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
            red = img[y][x][0] #mengambil nilai red dari pixel (y,x) pada gambar
            green = img[y][x][1] #mengambil nilai green dari pixel (y,x) pada gambar
            blue = img[y][x][2] #mengambil nilai blue dari pixel (y,x) pada gambar
            gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
            gray = int(thc * pow(gray, thy)) #operasi aritmatika thc dikalikan pow
            if gray > 255: #untuk mengecek nilai gray jika lebih dari nilai 255
                gray = 255 #jika lebih dari 255, maka nilai gray diset jadi 255
            if gray < 0: #untuk mengecek nilai gray jika kurang dari 0
                gray = 0 #jika kurang dari 0, maka nilai gray diset jadi 0
            img_nthpower[y][x] = (gray, gray, gray) #nilai keabuan yang telah diubah pada setiap pixel (y,x) disimpan pada variabel img_nthpower
```

```python
nthpower(50, 100) #menerapkan nthpower untuk parameter c nilainya 50 dan parameter y nilainya 100
plt.imshow(img_nthpower) #menampilkan gambar dari img_nthpower
plt.title("Nth Power") #diberi judul Nth Power
plt.show() #untuk menampilkan output di layar
```

```python
def nthrootpower(c, y): #mendefinisikan nthrootpower dengan parameter c dan y
    thc = c / 100 #inisialisasi nilai c dibagi 100 untuk variabel thc
    thy = y / 100 #inisialisasi nilai y dibagi 100 untuk vatiabel thy
    for y in range(0, img_height): #untuk melakukan loop pada setiap nilai y
        for x in range(0, img_width): #untuk melakukan loop pada setiap nilai x
            red = img[y][x][0] #mengambil nilai red dari pixel (y,x) pada gambar
            green = img[y][x][1] #mengambil nilai green dari pixel (y,x) pada gambar
            blue = img[y][x][2] #mengambil nilai blue dari pixel (y,x) pada gambar
            gray = (int(red) + int(green) + int(blue)) / 3 #menghitung nilai rata-rata dari nilai red, green, dan blue untuk mendapatkan nilai grayscale
            gray = int(thc * pow(gray, 1./thy)) #operasi aritmatika thc dikalikan pow
            if gray > 255: #untuk mengecek nilai gray jika lebih dari 255
                gray = 255 #jika lebih dari 255, maka nilai gray diset jadi 255
            if gray < 0: #untuk mengecek nilai gray jika kurang dari 0
                gray = 0 #jika kurang dari 0, maka nilai gray diset jadi 0
            img_nthpower[y][x] = (gray, gray, gray) #nilai keabuan yang telah diubah pada setiap pixel (y,x) disimpan pada variabel img_nthrootpower
```
